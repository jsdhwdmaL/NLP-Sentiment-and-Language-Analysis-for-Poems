## Preprocessing

```python
import string
string.punctuation
```

```python
punct_list = list(string.punctuation + '\n' + '\r')
def remove_punctuation(text):
    for punc in punct_list:
        if punc in text:
            text = text.replace(punc, ' ')
    return text.strip()
```

```python
rawData['clean_content'] = rawData['content'].apply(remove_punctuation)
# rawData.head()
```

```python
# Tokenize -- create a new tokenized column
import re

def tokenize(text):
    toks = re.split('\W+', text)
    return toks

rawData['tok_clean_content'] = rawData['clean_content'].apply(lambda x : tokenize(x.lower()))
# rawData.head()
```

```python
# Remove stopwords
from nltk.corpus import stopwords
stw = stopwords.words('english')
eliza_stw = ['adieu', 'anon', 'attend', 'aye', 'but soft', 'doth', "e'en", "e'er", 'haply', 'hence', 'hither', 'sirrah', 'thee', 'thither', 'thou', 'thou art', 'thy', 'whence', 'wilt', 'withal', 'hath']
#print(stw)

def re_stop(text):
    after = [word for word in text if word not in stw]
    after_two = [wording for wording in after if wording not in eliza_stw]
    return after_two

rawData['nostop_clean_content'] = rawData['tok_clean_content'].apply(lambda x : re_stop(x))
# rawData.head()
```

```python
# Lemmatize

#nltk.download('omw-1.4')
#nltk.download('wordnet')

wnl = nltk.WordNetLemmatizer()

def lemmatizing(text):
    after = [wnl.lemmatize(word) for word in text]
    return after

rawData['lemmatized_content'] = rawData['nostop_clean_content'].apply(lambda x : lemmatizing(x))
rawData.head(5)
# rawData.info()
```

```python
df_modern = rawData[rawData.age.isin(['Modern'])]
df_renaissance = rawData[rawData.age.isin(['Renaissance'])]
#(rawData[rawData.age.isin(['Renaissance'])]).to_csv('/Users/jojoli/Documents/夏校申请:项目制作/英才计划/正式培养/english/datasets/renaissance_tokenized.csv')
#(rawData[rawData.age.isin(['Modern'])]).to_csv('/Users/jojoli/Documents/夏校申请:项目制作/英才计划/正式培养/english/datasets/modern_tokenized.csv')
df_renaissance.head()
```

## Basic Imagery Extraction

### POS tagging & Noun Extraction

```python
from nltk import pos_tag
from nltk import word_tokenize

def check_noun(text):
    if text[1] == 'NN' or text[1] == 'NNP' or text[1] == 'NNPS' or text[1] == 'NNS':
        return True

def extract_noun(text):
    temp = ""
    for each in pos_tag(text): # POS tagging
        if (check_noun(each) == True):
            temp += each[0] + " " # extract nouns
    temp = temp.strip()
    return temp
```

```python
df_renaissance['all_nouns'] = df_renaissance['lemmatized_content'].apply(lambda x : extract_noun(x))
df_modern['all_nouns'] = df_modern['lemmatized_content'].apply(lambda x : extract_noun(x))
df_renaissance.head(5)
```

```python
df_renaissance['tok_all_nouns'] = df_renaissance['all_nouns'].apply(lambda x : tokenize(x.lower()))
df_modern['tok_all_nouns'] = df_modern['all_nouns'].apply(lambda x : tokenize(x.lower()))
#list(df_renaissance.tok_all_nouns.head(1))
df_renaissance.head()
```

### TF-IDF Keyword Extraction

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
cv = CountVectorizer()

def sort_coo(coo_matrix):
    tuples = zip(coo_matrix.col, coo_matrix.data)
    return sorted(tuples, key=lambda x: (x[1], x[0]), reverse=True)


def extract_topn_from_vector(feature_names, sorted_items, begin=0, end=10, lastbe=-10, lastend=-1):
    """get the feature names and tf-idf score of top n items"""

    # use only interval items from vector
    sorted_items = sorted_items[begin:end] + sorted_items[lastbe:lastend]
    score_vals = []
    feature_vals = []

    # word index and corresponding tf-idf score
    for idx, score in sorted_items:
        #keep track of feature name and its corresponding score
        score_vals.append(round(score, 3))
        feature_vals.append(feature_names[idx])

    # create a tuples of feature,score
    # results = zip(feature_vals,score_vals)
    results= {}
    for idx in range(len(feature_vals)):
        results[feature_vals[idx]]=score_vals[idx]

    return results


def tf_idf_calculator(enter_text, begin=0, end=10, lastbe=-10, lastend=-1):
    docs = np.array(enter_text)
    word_count_vector = cv.fit_transform(docs)
    print("Shape: ")
    print(word_count_vector.shape)

    # Convert the result to the TF-IDF matrix
    # Goal: apply rate values for tokens depending on how many times they appear
    tfidf = TfidfTransformer(smooth_idf=True,use_idf=True)
    tfidf.fit(word_count_vector)

    feature_names = cv.get_feature_names_out()
    doc = docs[0]
    for i in range (1, len(docs)):
        doc = doc + " " + docs[i]

    # generate tf-idf for the given document
    tf_idf_vector = tfidf.transform(cv.transform([doc]))

    # sort the tf-idf vectors by descending order of scores
    sorted_items = sort_coo(tf_idf_vector.tocoo())

    # extract only the top n; n here is 40
    keywords = extract_topn_from_vector(feature_names, sorted_items, begin, end, lastbe, lastend)

    # now print the results
    print("\n=====Doc=====")
    # print(doc)
    print("\n===Keywords===")
    for k in keywords:
        print(k, keywords[k])
```

```python
tf_idf_calculator(df_renaissance['all_nouns'], 5, 15, -30, -20)
```

```python
tf_idf_calculator(df_modern['all_nouns'], 5, 15, -35, -25)
```
